Considere a fun√ß√£o $$ y(x) = x^4 - 4x^3 + x^2 + 1$$ no intervalo [-2,4].


Determine dois intervalos, de amplitude no m√°ximo igual a 1, que contenham (cada um) uma  das 4 ra√≠zes da fun√ß√£o.

**RESPOSTA**

Vamos desenhar o gr√°fico da fun√ß√£o acima no intervalo de -2 at√© 4, incluindo a grade (para uma melhor visualiza√ß√£o).

```python
import matplotlib.pyplot as plt                   #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                # importamos a biblioteca numpy para lidar com lista de numeros
x = np.arange(-2,4.1,0.1)                         # x tem in√≠cio em -2, fim em 4, incremento de 0.1
y = x**4 - 4* x**3 + x**2 +1                      # funcao y (como definida acima)
plt.plot(x,y)                                     # desenhamos a funcao y
plt.grid()                                        # colocamos a grade
plt.xlabel("eixo x")                              # colocamos a legenda no eixo x
plt.ylabel("eixo y")                              # colocamos a legenda no eixo y
```

Olhando o gr√°fico podemos deduzir que a fun√ß√£o cruza a linha $y(x)=0$ nos intervalos $[0,1]$ (no ponto aproximadamente de $x=0.9$), e no intervalo $[3,4]$ (no ponto aproximadamente $x=3.6$).

Olhando o gr√°fico acima podemos deduzir que:

(a) Resposta para item (a) √© 2.

Com o objetivo de aumentar a produ√ß√£o de alimentos em certa regi√£o, uma secretaria de agricultura encomendou a uma equipe de agr√¥nomos um estudo sobre as potencialidades do solo dessa regi√£o. Na an√°lise da temperatura do solo, a equipe efetuou medi√ß√µes di√°rias, durante quatro dias consecutivos, em intervalos de uma hora.

As medi√ß√µes tiveram in√≠cio √†s 6 horas da manh√£ do primeiro dia. Os estudos indicaram que a temperatura $y$ (medida em graus Celsius), e o tempo $x$ (representando o n√∫mero de horas decorridas ap√≥s o in√≠cio das observa√ß√µes), relacionavam-se atrav√©s da express√£o:

$$y(x)= 26 + 5 \cos(\frac{\pi}{12}  x + \frac{4 \pi}{3}).$$

Desenhe o gr√°fico acima, e com base nesse gr√°fico, identifique as afirmativas corretas:

(a) A temperatura do solo, √†s 6 horas da manh√£ do primeiro dia, foi de aproximadamente 23,5 ¬∫C.


(b) A fun√ß√£o $y(x)$ atinge valor m√°ximo igual a 30 ¬∫C.

(c) A temperatura do solo atingiu o valor m√°ximo, no primeiro dia, √†s 14 h.

(d) A fun√ß√£o $y(x)$ √© crescente no intervalo [0,8].


**RESPOSTA**

Qual intervalo de tempo (valores no eixo $x$) devemos considerar? Como a equipe efetuou medi√ß√µes di√°rias, **durante quatro dias consecutivos, em intervalos de uma hora**, temos que o valor de $x$ deve variar de 0 at√© $(4 * 24)$ = 96.

Vamos desenhar o gr√°fico da fun√ß√£o acima nesse intervalo, de 0 at√© 96, incluindo a grade (para uma melhor visualiza√ß√£o).

```python
import matplotlib.pyplot as plt                     #importamos a biblioteca matplotlib.pyplot para graficos
%matplotlib inline
import numpy as np                                  # importamos a biblioteca numpy para lidar com lista de numeros
import math                                         # importamos a biblioteca math para poder usar math.pi (ou seja, para numero Pi)
x = np.arange(0,96.1,0.1)                           # x tem in√≠cio em 0, fim em 96
y = 26 + 5 * np.cos((math.pi/12)*x + 4*math.pi/3)   # funcao y (como definida acima)
plt.plot(x,y)                                       # desenhamos a funcao
plt.grid()                                          # colocamos a grade
plt.xlabel("tempo (x)")                             # colocamos a legenda no eixo x
plt.ylabel("temperatura (y)")                       # colocamos a legenda no eixo y
```

Observando o gr√°fico acima, podemos concluir que a fun√ß√£o $y(x)$ atinge m√°ximo 4 vezes, e que esse m√°ximo √© valor maior que 30 ¬∫C. 

Logo, a afirmativa (b) √© FALSA.

Vamos olhar a afirmativa (a). Note que 6 horas da manh√£ do primeiro dia, quando as medi√ß√µes tiveram in√≠cio, corresponde ao valor $x=0$ no nosso gr√°fico (pois $x$ representa o n√∫mero de horas decorridas ap√≥s o in√≠cio das observa√ß√µes - como definido no exerc√≠cio).

Similarmente, na afirmativa (c), 14h de primeiro dia corresponde ao valor $x=14-6=8$ no nosso gr√°fico.

Afirmativa (d) pede para analisar se o gr√°fico √© crescente no intervalo [0,8].

Para descobrir se essas afirmativas s√£o verdadeiras ou falsas, devemos olhar com mais precis√£o o in√≠cio do gr√°fico, digamos no intervalo [0,10]. Vamos repetir o c√≥digo acima, s√≥ que n√£o precisamos incluir as bibliotecas de novo; e neste c√≥digo, $x$ vai ter in√≠cio em 0 e fim em 10.



```python
x = np.arange(0,10.1,0.1)                        # x tem in√≠cio em 0, fim em 10
y = 26 + 5 * np.cos(math.pi/12*x + 4*math.pi/3)   
plt.plot(x,y)                  
plt.grid()            
plt.xlabel("tempo (x)")             
plt.ylabel("temperatura (y)")       
```


Olhando o gr√°fico acima, podemos afirmar que:
- A temperatura do solo, √†s 6 horas da manh√£ do primeiro dia (ou seja, no ponto $x=0$), foi de aproximadamente 23,5 ¬∫C. Logo (a) √© verdadeiro
- A temperatura do solo (valor de $y(x)$) atingiu o valor m√°ximo no ponto $x=8$, ou seja, √†s 6+8 = 14 h do primeiro dia. Logo, (c) √© verdadeiro.
- A fun√ß√£o $y(x)$  cresce de $x=0$ at√© $x=8$ (quando $x=8$ atinge m√°ximo), e depois come√ßa a decrescer. Logo, (d) √© verdadeiro tamb√©m.
 
Observe que  a fun√ß√£o ùë¶(ùë•) de fato n√£o atinge valor m√°ximo igual a 30 ¬∫C.  No gr√°fico acima podemos ver que o valor m√°ximo da fun√ß√£o ùë¶(ùë•) √© igual a 31 ¬∫C.

Podemos desenhar fun√ß√µes de itens (a)-(c) num gr√°fico e fun√ß√µes de itens (d) e (e) no outro.
